```python
Q=[]
while True:
    print("1. Insert   2. Delete   3. Display   4. Exit")
    ch=int(input("Enter choice: "))
    if ch==1:
        bid=int(input("Enter book id: "))
        bname=input("Enter book name: ")
        aut=input("Enter author name: ")
        Q.append([bid,bname,aut])
    elif ch==2:
        if Q==[]:
            print("Queue is empty...")
        else:
            bid,bname,aut=Q.pop(0)
            print("Deleted book",bid,bname)
    elif ch==3:
        if Q==[]:
            print("Queue is empty...")
        else:
            for i in range(len(Q)):
                if i==0:
                    print(Q[i],"<----First value")
                else:
                    print(Q[i])
    elif ch==4:
        break
    else:
        print("Invalid Input!!\nEnter choice again...")
    print()
```
