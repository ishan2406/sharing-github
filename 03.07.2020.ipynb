```python
s=[]
while True:
    print("1. Push\t 2. Pop\t 3. Display\t 4. Exit")
    ch=int(input("Enter choice: "))
    if ch==1:
        ele=int(input("Enter no. to add in stack: "))
        s.append(ele)
        print()
    elif ch==2:
        if s==[]:
            print("Stack is empty")
        else:
            print("Value deleted",s.pop())
            print()
    elif ch==3:
        l=len(s)
        while l>0:
            print(s[l-1])
            l=l-1
        print()
    else:
        break
```

```python
E=[]
while True:
    print("1. Push\t 2. Pop\t 3. Display\t 4. Exit")
    ch=int(input("Enter choice: "))
    if ch==1:
        eid=int(input("Enter Emp Id:"))
        ename=input("Enter Emp Name: ")
        E.append((eid,ename))
        print()
    elif ch==2:
        if E==[]:
            print("Stack is empty")
        else:
            eid,ename=E.pop()
            print("Employee deleted:",eid,ename)
            print()
    elif ch==3:
        l=len(E)
        while l>0:
            print(E[l-1])
            l=l-1
        print()
    else:
        break
```
