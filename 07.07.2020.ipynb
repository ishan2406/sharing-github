```python
S=[]
while True:
    print("1. Push   2. Pop   3. Display   4. Exit")
    ch=int(input("Enter choice: "))
    if ch==1:
        bid=int(input("Enter book id:"))
        bname=input("Enter book name: ")
        aut=input("Enter author name:")
        S.append([bid,bname,aut])
        print()
    elif ch==2:
        if S==[]:
            print("Stack is empty")
        else:
            bid,bname,aut=S.pop()
            print("Book deleted:",bid,bname,aut)
            print()
    elif ch==3:
        l=len(S)
        while l>0:
            print(S[l-1])
            l=l-1
        print()
    else:
        break
```
