```python
def isEmpty(st):
    if st==[]:
        return True
    else:
        return False
def Push(st,ele):
    st.append(ele)
    top=len(st)-1
def Pop(st):
    if isEmpty(st):
        return "Underflow"
    else:
        ele=st.pop()
        if len(st)==0:
            top=None
        else:
            top=len(st)-1
        return ele
def Peek(st):
    if isEmpty(st):
        return "Underflow"
    else:
        top=len(st)-1
        return st[top]
def Display(st):
    if isEmpty(st):
        return "Underflow"
    else:
        top=len(st)-1
        print(st[top],"<----Top")
        for i in range(top-1,-1,-1):
            print(st[i])
stack=[]
while True:
    print("1.Push   2.Pop   3.Peek   4.Display   5.Exit")
    ch=int(input("Enter choice: "))
    if ch==1:
        ele=int(input("Enter element to be inserted: "))
        Push(stack,ele)
    elif ch==2:
        ele=Pop(stack)
        if ele=="Underflow":
            print("Underflow!!\nStack is empty...")
        else:
            print("Popped element is",ele)
    elif ch==3:
        ele=Peek(stack)
        if ele=="Underflow":
            print("Underflow!!\nStack is empty...")
        else:
            print("Topmost element is",ele)
    elif ch==4:
        if isEmpty(stack):
            print("Underflow!!\nStack is empty...")
        else:
            Display(stack)
    elif ch==5:
        break
    else:
        print("Invalid choice!!\nEnter choice again...")
    print()
```
