```python
import matplotlib.pyplot as pl
a=["Math","Phy","Chem","Comp","Eng"]
b=[76,80,69,89,80]
pl.xlabel("Subjects")
pl.ylabel("Marks")
pl.plot(a,b,"g")    #default light blue colour of line, b-dark blue, k-black, g-green, m-magenta, r-red
```

```python
import numpy as np
x=np.arange(1,20,3)
x
```

```python
x=np.arange(0.,10,0.1)
a=np.cos(x)
b=np.sin(x)
pl.xlabel("Cos of x")
pl.ylabel("Sin of x")
pl.plot(x,a,"g",linewidth=5)
pl.plot(x,b,"m",linewidth=6,linestyle="dotted")    #linestyles - dashed, dashdot, dotted
```

```python
p=[1,2,3,4]
q=[2,4,6,8]
pl.plot(p,q,"r+",marker=">",markersize=5,markeredgecolor="blue")    #Markers - d, D, o, *, +, x, 1, 2, >, <
#r+ makes the line of the graph disappear
```

```python
a=np.arange(1,20,1.25)
b=np.log(a)
c=np.cos(a)
pl.plot(a,b,"r+",marker="*",markeredgecolor="green")
pl.plot(a,c,"y",linestyle="dashdot",linewidth=2,marker="s",markersize=10,markeredgecolor="red")
```

```python
import matplotlib.pyplot as pl
a,b,c=[1,2,3,4,5],[2,4,6,8,10],[1,4,9,16,25]
pl.bar(a,b,width=0.5,color="r")
pl.xlabel("Values")
pl.ylabel("Double of values")
```

```python
a=["Delhi","Mumbai","Chennai","Ahmedabad"]
b=[12435678,46751342,24659745,54132745]
pl.barh(a,b,color=["r","g","b","m"])    #Cannot use width in horizontal bars
```

```python
import numpy as np
A=[2,4,6,8]
B=[2.8,3.5,6.5,4.1]
X=np.arange(len(A))
pl.bar(X,A,color="g",width=0.3)
pl.bar(X+0.3,B,color="m",width=0.3)
```

```python
Marks=[[65,67,88],[89,90,99],[14,34,45]]
X=np.arange(len(Marks))
pl.ylabel("Marks")
pl.bar(X+0.0,Marks[0],color="k",width=0.2)
pl.bar(X+0.2,Marks[1],color="r",width=0.2)
pl.bar(X+0.4,Marks[2],color="g",width=0.2)
```
